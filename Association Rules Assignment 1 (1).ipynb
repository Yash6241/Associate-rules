```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv(r"C:\Users\Yash Bhosale\Downloads\book (4).csv")
df
```

```python
frequent_item = apriori(df,min_support=0.1, use_colnames= True)
frequent_item
```

```python
fi = apriori(df, min_support= 0.09, use_colnames= True)
fi
```

```python
rules = association_rules(frequent_item, metric = 'lift', min_threshold= 0.7)
rules
```

```python
r = association_rules(fi, metric= 'lift', min_threshold= 0.6)
r
```

```python
rules.sort_values('lift', ascending= False)[0:20]
```

```python
rr = rules[rules.lift > 1]
rr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.regplot(x = rules.support, y = rules.confidence)
```

```python
sns.regplot(x = rules.support, y = rules.lift)
```

```python
fit = np.polyfit(rules['lift'], rules['confidence'], 1)
fit_fn = np.poly1d(fit)
plt.plot(rules['lift'], rules['confidence'], 'yo', rules['lift'], 
 fit_fn(rules['lift']))
```
